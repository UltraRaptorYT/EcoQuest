```python
# # PART 1: DETECTING IMAGE

# import os
# import requests
# from dotenv import load_dotenv

# # Load environment variables
# load_dotenv()

# # Get API key from environment variable
# apikey = os.getenv('OPENAI_KEY')

# from openai import OpenAI 
# import base64
# import requests

# # Function to encode the image
# def encode_image(image_path):
#   with open(image_path, "rb") as image_file:
#     return base64.b64encode(image_file.read()).decode('utf-8')

# # Path to your image
# image_path = "cardboard.jpeg"

# # Getting the base64 string
# base64_image = encode_image(image_path)

# headers = {
#   "Content-Type": "application/json",
#   "Authorization": f"Bearer {apikey}"
# }
# payload = {
#     "model": "gpt-4-vision-preview",
#     "messages": [
#         {
#             "role": "system",
#             "content": "Only tell me what material and item name in the image, in the (material, item) format"
#         },
#         {
#             "role": "user",
#             "content": [
#                 {"type": "text", "text": "What’s in this image?"},
#                 {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}},
#             ],
#         },
#     ],
#     "max_tokens": 20
# }

# response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

# # Assuming 'response' contains the API response
# response_json = response.json()

# # Extracting the content from the response
# image_detection_output = response_json['choices'][0]['message']['content']

# # Printing the content
# print(image_detection_output)

import os
import base64
import requests
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Get API key from environment variable
apikey = os.getenv('OPENAI_KEY')

# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# Function to create the payload
def create_payload(base64_image):
    return {
        "model": "gpt-4-vision-preview",
        "messages": [
            {
                "role": "system",
                "content": "Only tell me what material and item name in the image, in the (material, item) format"
            },
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": "What’s in this image?"},
                    {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}},
                ],
            },
        ],
        "max_tokens": 20
    }

# Function to get the response from the API
def get_image_detection_response(image_path):
    base64_image = encode_image(image_path)
    payload = create_payload(base64_image)
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {apikey}"
    }
    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
    return response.json()

# Function to extract the content from the response
def extract_image_detection_output(response_json):
    return response_json['choices'][0]['message']['content']

# Main function to run the detection
def main_image_detect(image_path):
    response_json = get_image_detection_response(image_path)
    image_detection_output = extract_image_detection_output(response_json)
    print(image_detection_output)
    return image_detection_output

# Path to your image
image_path = "coke_can.png"

material_item = main_image_detect(image_path)
```
